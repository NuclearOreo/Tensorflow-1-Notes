# MNIST CNN

```python
#Importing Tensorflow
import tensorflow as tf
```

```python
#Importing input_data function to get the MNIST Dataset
from tensorflow.examples.tutorials.mnist import input_data
```

```python
#Grab the MNIST Dataset
mnist = input_data.read_data_sets('MNIST_data', one_hot=True)
```

```python
#Initialize weights for layer
def init_weights(shape):
    init_random_dist = tf.truncated_normal(shape=shape, stddev=0.1)
    return tf.Variable(init_random_dist)
```

```python
#Initialize Bias for layer
def init_bias(shape):
    init_bias_vals = tf.constant(0.1,shape=shape)
    return tf.Variable(init_bias_vals)
```

```python
#Hepler function to perform a convolution
def conv2d(x,W):
    return tf.nn.conv2d(x,W,strides=[1,1,1,1], padding='SAME')
```

```python
#Create a max pooling layer
def max_pooling_2by2(x):
    return tf.nn.max_pool(x,strides=[1,2,2,1], ksize=[1,2,2,1], padding='SAME')
```

```python
#Helper function to create a convolutional layer
def convolutional_layer(input_x,shape):
    W = init_weights(shape)
    b = init_bias([shape[3]])
    return tf.nn.relu(conv2d(input_x,W)+b)
```

```python
#Helper functioncreate fully connected layer
def normal_full_layer(input_layer, size):
    input_size = int(input_layer.get_shape()[1])
    W = init_weights([input_size,size])
    b = init_bias([size])
    return tf.matmul(input_layer,W)+b
```

```python
#Input placeholder
x = tf.placeholder(tf.float32,[None, 784])
```

```python
#Output placeholder
y_true = tf.placeholder(tf.float32, [None, 10])
```

```python
#Reshape the image from 784 to 28 by 28
x_image = tf.reshape(x, [-1,28,28,1])
```

```python
#Create a convolutional layer with max pooling
convo_1 = convolutional_layer(x_image, shape=[5,5,1,32])
convo_1_pooling = max_pooling_2by2(convo_1)
```

```python
#Creating another convolutional layer with max pooling
convo_2 = convolutional_layer(convo_1_pooling, shape=[5,5,32,64])
convo_2_pooling = max_pooling_2by2(convo_2)
```

```python
#Flatten the output from conv_2_pooling and pass it into a fully connected layer
convo_2_flat = tf.reshape(convo_2_pooling, [-1,7*7*64])
full_layer_one = tf.nn.relu(normal_full_layer(convo_2_flat,1024))
```

```python
#Create a dropout layer
hold_prob = tf.placeholder(tf.float32)
full_one_dropout = tf.nn.dropout(full_layer_one, rate=hold_prob)
```

```python
#Get the predictions
y_pred = normal_full_layer(full_layer_one, 10)
```

```python
cross_entropy = tf.reduce_mean(tf.nn.softmax_cross_entropy_with_logits_v2(labels=y_true, logits=y_pred))
```

```python
#Create an optimizer and a trainer
optimizer = tf.train.AdamOptimizer(learning_rate=0.0001)
train = optimizer.minimize(cross_entropy)
```

```python
#Variable to create an init function
init = tf.global_variables_initializer()
```

```python
#Run Session with 5000 epoch
steps = 10000

with tf.Session() as sess:
    
    sess.run(init)
    
    for i in range(steps):
        
        batch_x, batch_y = mnist.train.next_batch(50)
        
        sess.run(train, feed_dict={x:batch_x, y_true: batch_y, hold_prob: 0.5})
        
        if i%100 == 0:
            
            print('ON STEP: {}'.format(i))
            print('ACCURACY: ')
            
            matches = tf.equal(tf.argmax(y_pred,1),tf.argmax(y_true,1))
            
            acc = tf.reduce_mean(tf.cast(matches,tf.float32))
            
            print(sess.run(acc,feed_dict={x: mnist.test.images, y_true: mnist.test.labels, hold_prob:1.0}))
            print('\n')
```
